# **MNIST Dataset Classification Using Dimensionality Reduction Methods**

## Load the MNIST dataset and split it into a training set and a test set (take the first 60,000 instances for training, and the remaining 10,000 for testing).

```python
# YOUR CODE HERE

# Global function
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, LocallyLinearEmbedding, MDS
```

```python
# Load the data
df = pd.read_csv("/content/mnist_dataset.csv", index_col=0)
df
```

```python
# Identify the target and the features
TARGET="label"
columns_predictors = [col for col in df.columns if col not in [TARGET]]

print(TARGET)
print(columns_predictors)
```

```python
# Split data into a training set (60,000) and a test set (10,000)
df_train = df[:60000]
df_test = df[60000:]

print(df_train.shape, df_test.shape)
```

```python
# Define the X_train, y_train, X_test, and y_test
X_train = df_train[columns_predictors]
y_train = df_train[TARGET]
X_test = df_test[columns_predictors]
y_test = df_test[TARGET]

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# Scale the X training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
display(X_train_scaled.shape)
print(X_train_scaled)

# Scale the X testing data
X_test_scaled = scaler.fit_transform(X_test)
display(X_test_scaled.shape)
print(X_test_scaled)
```

## Train a Random Forest classifier on the dataset and time how long it takes, then evaluate the resulting model on the test set.

```python
%%time
# YOUR CODE HERE

# Train a Random Forest classifier
rfc_1 = RandomForestClassifier(n_estimators=50, random_state=42)
rfc_1.fit(X_train_scaled, y_train)
```

```python
# Evaluate the resulting model on the test
y_pred = rfc_1.predict(X_test_scaled)

# Performance evaluation
print("Performance Evaluation:\n")
print("Accuracy Score:\n", accuracy_score(y_test, y_pred), "\n")
print("Precision:\n", precision_score(y_test, y_pred, average='macro'), "\n")
print("Recall:\n", recall_score(y_test, y_pred, average='macro'), "\n")
print("F1 Score:\n", f1_score(y_test, y_pred, average='macro'), "\n")
print("Roc Auc Score:\n", roc_auc_score(y_test, rfc_1.predict_proba(X_test_scaled), multi_class='ovr'), "\n")
```

## Use PCA to reduce the dataset’s dimensionality, with an explained variance ratio of 95%. Train a new Random Forest classifier on the reduced dataset and see how long it takes. Was training much faster? Next evaluate the classifier on the test set: how does it compare to the previous classifier?

```python
# YOUR CODE HERE
%%time
# Use PCA to reduce the dataset’s dimensionality
pca = PCA(n_components=0.95)
X_train_reduced = pca.fit_transform(X_train_scaled)
X_train_reduced.shape
```

```python
%%time
# Train a New Random Forest classifier using the reduced X_train
rfc_2 = RandomForestClassifier(n_estimators=50, random_state=42)
rfc_2.fit(X_train_reduced, y_train)
```

```python
%%time
# Evaluate the classifier on the test set
X_test_reduced = pca.transform(X_test_scaled)
y_pred = rfc_2.predict(X_test_reduced)
```

```python
# Performance evaluation
print("Performance Evaluation:\n")
print("Accuracy Score:\n", accuracy_score(y_test, y_pred), "\n")
print("Precision:\n", precision_score(y_test, y_pred, average='macro'), "\n")
print("Recall:\n", recall_score(y_test, y_pred, average='macro'), "\n")
print("F1 Score:\n", f1_score(y_test, y_pred, average='macro'), "\n")
print("Roc Auc Score:\n", roc_auc_score(y_test, rfc_2.predict_proba(X_test_reduced), multi_class='ovr'), "\n")
```

**Comparison:**

Applying PCA actually slowed down the training process, extending it by more than 1 minute compared to training without PCA.

The performance deteriorated after using PCA, likely due to the loss of important dimensions that were critical for the classifier's accuracy.

## Use t-SNE to reduce the MNIST dataset, show result graphically.

```python
# YOUR CODE HERE

# Ploting the graphs takes too much time, so we subsample the data to only 5000 data
n_samples = 5000
indices = np.random.permutation(60000)[:n_samples]
X_subsampled_data = X_train_scaled[indices]
y_subsampled_data = y_train[indices]
```

```python
%%time
# Use t-SNE to reduce the MNIST dataset, show result graphically
reducer = TSNE(n_components=2, n_jobs=-1)
X_reduced = reducer.fit_transform(X_subsampled_data)
display(X_reduced.shape)
display(X_reduced)
fig = px.scatter(x=X_reduced[:,0], y=X_reduced[:,1], color=y_subsampled_data, title="TSNE")
fig.update_layout(coloraxis_colorbar=dict(title=TARGET), height=1000)
fig.update_yaxes(scaleanchor = "x", scaleratio = 1)
fig.show()
```

## Compare with other dimensionality methods: _Locally Linear Embedding_ (LLE) or _Multidimensional scaling_ (MDS).

```python
%%time
# YOR CODE HERE

#  Use LLE to reduce the MNIST dataset, show result graphically
reducer = LocallyLinearEmbedding(n_components=2, n_jobs=-1)
X_reduced = reducer.fit_transform(X_subsampled_data)
display(X_reduced.shape)
display(X_reduced)
fig = px.scatter(x=X_reduced[:,0], y=X_reduced[:,1], color=y_subsampled_data, title="LLE")
fig.update_layout(coloraxis_colorbar=dict(title=TARGET), height=1000)
fig.update_yaxes(scaleanchor = "x", scaleratio = 1)
fig.show()
```

```python
%%time
#  Use MDS to reduce the MNIST dataset, show result graphically
reducer = MDS(n_components=2, n_jobs=-1)
X_reduced = reducer.fit_transform(X_subsampled_data)
display(X_reduced.shape)
display(X_reduced)
fig = px.scatter(x=X_reduced[:,0], y=X_reduced[:,1], color=y_subsampled_data, title="MDS")
fig.update_layout(coloraxis_colorbar=dict(title=TARGET), height=1000)
fig.update_yaxes(scaleanchor = "x", scaleratio = 1)
fig.show()
```

**Comparison:**
 - t-SNE: Exhibits clear cluster separation, indicating effective dimensionality reduction and distinct groupings within the data.
 - LLE: Less effective than t-SNE because it looks like two broad arms extending from a denser cluster at the center, and there is data overlap.
 - MDS: Better than LLE but not as clear as t-SNE because it appears data overlap.
